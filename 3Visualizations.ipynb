## Exploratory Visualizations
- Look at distributions.
- Look at correlations.
- Look at relationships to target (scatter plots for continuous, box plots for categorical).

```python
plt.figure(figsize=(10,7))
plt.bar('Overall Cond', 'SalePrice', data=trainall,color='coral')
plt.ylabel('Sale Price',size=20)
plt.xlabel('Overall Condition',size=20)
#x_ticks=np.arange(0,6000,500)
#plt.xticks(x_ticks)


plt.title('Sale Price vs Overall Condition',size=25)
;
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x='sq_ft', y='SalePrice',hue='Full Bath',alpha=.8,size='sq_ft',sizes=(15, 200),x_bins=200,
                    data=trainall,
                    #scale='width', 
            palette='Set1')
plt.ylabel('Sale Price',size=15)
plt.xlabel('Total Square Footage',size=15)
plt.title('Sale Price vs Total Square Footage : Colored by Number of Full Bathrooms',size=20);
```

```python
plt.figure(figsize=(10,7))
sns.violinplot(x='Overall Qual', y='SalePrice',
                    data=trainall,
                    scale='width', palette='Set3')


plt.ylabel('Sale Price',size=20)
plt.xlabel('Overall Quality',size=20)

plt.title('Sale Price vs Overall Quality',size=30);
```

```python
plt.figure(figsize=(10,7))
sns.swarmplot(x='Overall Cond', y='SalePrice', data=trainall,hue='SalePrice')
plt.ylabel('Sale Price',size=20)
plt.xlabel('Overall Cond',size=20)
plt.legend().set_visible(False)
plt.title('Sale Price vs Overall Condition',size=30);
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x='Total Bsmt SF', y='SalePrice'
              ,data=trainall,hue='CemntBd'
              ,alpha=.8
              ,size='Overall Qual'
              ,sizes=(15, 200)
              ,x_bins=200
                    
                    #,scale='width' 
            ,palette='Set1'
             )
plt.ylabel('Sale Price',size=15)
plt.xlabel('Total Basement Square Footage',size=15)
plt.title('Sale Price vs Total Basement Square Footage : Colored by Cement Siding',size=20);
```

```python

plt.figure(figsize=(10,7))
sns.violinplot(x='Full Bath', y='SalePrice',
                    data=trainall,legend_out=True,
                    scale='width', palette='Set3')
sns.swarmplot(x='Half Bath', y='SalePrice', data=trainall)

plt.ylabel('Sale Price',size=20)
plt.xlabel('Bathrooms',size=20)
plt.legend().set_visible(True)

plt.title('Sale Price vs Full and Half Bathrooms',size=25);
```

```python
sns.catplot(data=trainall, x='Street', y='SalePrice', 
    hue='SalePrice', height=8, 
            aspect=11.7/8.27,
            legend_out=False
           )

plt.ylabel('Sale Price',size=20)
plt.xlabel('Driveway Type',size=20)
plt.legend().set_visible(False)
plt.title('Sale Price vs Driveway Type',size=25)
;
```

```python
sns.catplot(x='Overall Cond',y='SalePrice',kind='box',data=trainall);
```

```python

plt.figure(figsize=(10,7))
sns.swarmplot(x='age', y='SalePrice', data=trainall,hue='SalePrice')
plt.ylabel('Sale Price',size=15)
plt.xlabel('Age',size=15)
plt.legend().set_visible(False)
plt.title('Sale Price vs Age of the House',size=18);
```

```python
plt.figure(figsize=(10,7))
sns.swarmplot(x='BsmtFin Type 1', y='SalePrice', data=trainall,hue='SalePrice')
plt.ylabel('Sale Price',size=15)
plt.xlabel('Condition of Basement',size=15)
plt.legend().set_visible(False)
plt.title('Sale Price vs Basement Finished Type',size=18);
```
